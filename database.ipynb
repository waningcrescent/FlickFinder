```python
from sqlalchemy.exc import IntegrityError
from sqlalchemy import text

def populate_movies_table(connection, df):
    # SQL queries
    check_query = text("SELECT 1 FROM movies WHERE imdb_id = :imdb_id")
    insert_query = text("""
        INSERT INTO movies (
            title, type, description, runtime, production_countries, imdb_id, 
            imdb_score, is_amazon, year, is_hotstar, director, rating, 
            is_netflix, poster_url
        ) VALUES (
            :title, :type, :description, :runtime, :production_countries, :imdb_id, 
            :imdb_score, :is_amazon, :year, :is_hotstar, :director, :rating, 
            :is_netflix, :poster_url
        )
    """)

    for _, row in df.iterrows():
        # Prepare movie data
        movie_data = {
            'title': row['title'] if pd.notna(row['title']) else None,
            'type': row['type'] if pd.notna(row['type']) else None,
            'description': row['description'] if pd.notna(row['description']) else None,
            'runtime': int(row['runtime']) if pd.notna(row['runtime']) else None,
            'production_countries': row['production_countries'] if pd.notna(row['production_countries']) else None,
            'imdb_id': row['imdb_id'] if pd.notna(row['imdb_id']) else None,
            'imdb_score': float(row['imdb_score']) if pd.notna(row['imdb_score']) else None,
            'is_amazon': int(row['is_amazon']) if pd.notna(row['is_amazon']) else 0,
            'year': int(row['year']) if pd.notna(row['year']) else None,
            'is_hotstar': int(row['is_hotstar']) if pd.notna(row['is_hotstar']) else 0,
            'director': row['director'] if pd.notna(row['director']) else None,
            'rating': row['rating'] if pd.notna(row['rating']) else None,
            'is_netflix': int(row['is_netflix']) if pd.notna(row['is_netflix']) else 0,
            'poster_url': row['poster_url'] if pd.notna(row['poster_url']) else None,
        }

        try:
            # Check if the movie already exists based on imdb_id
            exists = connection.execute(check_query, {"imdb_id": movie_data['imdb_id']}).fetchone()
            if exists:
                print(f"Skipping duplicate movie: {movie_data['title']} (IMDB ID: {movie_data['imdb_id']})")
                continue

            # Insert the movie if it doesn't exist
            connection.execute(insert_query, movie_data)
        except IntegrityError as e:
            print(f"Integrity error for row {movie_data['title']}: {e}")

    print("Movies table populated successfully (with duplicates skipped).")

populate_movies_table(connection, csv)
```

```python
from sqlalchemy import text
import pandas as pd

# Populate Genres Table
def populate_genres_table(connection, csv):
    # Extract unique genres from the 'genres' column
    unique_genres = set(
        genre.strip()
        for genres in csv['genres'].dropna()
        for genre in genres.split(',')
    )
    # Define the SQL query
    query = text("INSERT IGNORE INTO genres (name) VALUES (:name)")
    # Insert each unique genre
    for genre in unique_genres:
        connection.execute(query, {"name": genre})
    print("Genres table populated successfully")

# Populate Movie-Genres Table
def populate_movie_genres_table(connection, csv):
    # Queries to fetch movie_id and genre_id
    movie_query = text("SELECT id FROM movies WHERE title = :title")
    genre_query = text("SELECT id FROM genres WHERE name = :name")
    check_movie_genre_query = text(
        "SELECT 1 FROM movie_genres WHERE movie_id = :movie_id AND genre_id = :genre_id"
    )
    insert_movie_genre_query = text(
        "INSERT INTO movie_genres (movie_id, genre_id) VALUES (:movie_id, :genre_id)"
    )

    for _, row in csv.iterrows():
        if pd.notna(row['genres']):
            # Fetch movie_id using the movie title
            result = connection.execute(movie_query, {"title": row['title']}).fetchone()
            if result:
                movie_id = result[0]  # Access id using integer index
                genres = row['genres'].split(',')
                for genre in genres:
                    genre = genre.strip()
                    # Fetch genre_id for each genre
                    genre_result = connection.execute(genre_query, {"name": genre}).fetchone()
                    if genre_result:
                        genre_id = genre_result[0]  # Access id using integer index
                        # Check if this pair already exists
                        exists = connection.execute(
                            check_movie_genre_query,
                            {"movie_id": movie_id, "genre_id": genre_id},
                        ).fetchone()
                        if not exists:
                            # Insert the relationship into movie_genres
                            connection.execute(
                                insert_movie_genre_query,
                                {"movie_id": movie_id, "genre_id": genre_id},
                            )
    print("Movie genres table populated successfully")


# Populate Cast Table
def populate_cast_table(connection, csv):
    # Extract unique cast members from the 'cast' column
    unique_cast = set(
        member.strip()
        for cast_list in csv['cast'].dropna()
        for member in cast_list.split(',')
    )
    # Define the SQL query
    query = text("INSERT IGNORE INTO cast (name) VALUES (:name)")
    # Insert each unique cast member
    for cast_member in unique_cast:
        connection.execute(query, {"name": cast_member})
    print("Cast table populated successfully")

# Populate Movie-Cast Table
def populate_movie_cast_table(connection, csv):
    # Queries to fetch movie_id and cast_id
    movie_query = text("SELECT id FROM movies WHERE title = :title")
    cast_query = text("SELECT id FROM cast WHERE name = :name")
    check_movie_cast_query = text(
        "SELECT 1 FROM movie_cast WHERE movie_id = :movie_id AND cast_id = :cast_id"
    )
    insert_movie_cast_query = text(
        "INSERT INTO movie_cast (movie_id, cast_id) VALUES (:movie_id, :cast_id)"
    )

    for _, row in csv.iterrows():
        if pd.notna(row['cast']):
            # Fetch movie_id using the movie title
            result = connection.execute(movie_query, {"title": row['title']}).fetchone()
            if result:
                movie_id = result[0]  # Access id using integer index
                cast_members = row['cast'].split(',')
                for member in cast_members:
                    member = member.strip()
                    # Fetch cast_id for each cast member
                    cast_result = connection.execute(cast_query, {"name": member}).fetchone()
                    if cast_result:
                        cast_id = cast_result[0]  # Access id using integer index
                        # Check if this pair already exists
                        exists = connection.execute(
                            check_movie_cast_query,
                            {"movie_id": movie_id, "cast_id": cast_id},
                        ).fetchone()
                        if not exists:
                            # Insert the relationship into movie_cast
                            connection.execute(
                                insert_movie_cast_query,
                                {"movie_id": movie_id, "cast_id": cast_id},
                            )
    print("Movie cast table populated successfully")



populate_genres_table(connection, csv)         # Populate genres table
populate_movie_genres_table(connection, csv)  # Populate movie_genres table
populate_cast_table(connection, csv)          # Populate cast table
populate_movie_cast_table(connection, csv)    # Populate movie_cast table
```

```python
from sqlalchemy import text

def display_table_data(connection, table_name, conditions=None, limit=10):
    """
    Display data from a table with optional conditions.
    
    Parameters:
    - connection: SQLAlchemy connection object
    - table_name: str, name of the table to query
    - conditions: str, optional SQL conditions (e.g., "WHERE release_year = 2020")
    - limit: int, number of rows to display
    """
    base_query = f"SELECT * FROM {table_name} "
    if conditions:
        base_query += f" {conditions} "
    base_query += "LIMIT :limit"
    
    query = text(base_query)
    result = connection.execute(query, {"limit": limit})
    print(f"Data in {table_name} (showing up to {limit} rows):")
    for row in result:
        print(row)
    print("\n" + "="*50 + "\n")  # Separator for readability
```

```python
def get_movies_by_genre(connection, genre_name):
    query = """
    SELECT m.title, m.year AS release_year, m.imdb_score
    FROM movies m
    JOIN movie_genres mg ON m.id = mg.movie_id
    JOIN genres g ON mg.genre_id = g.id
    WHERE g.name = :genre_name
    """
    result = connection.execute(text(query), {"genre_name": genre_name})
    return result.fetchall()
```

```python
def get_movies_by_platform(connection, platform_name):
    query = """
    SELECT m.title, m.year AS release_year, m.rating
    FROM movies m
    WHERE 
        (:platform_name = 'Amazon' AND m.is_amazon = 1)
        OR (:platform_name = 'Netflix' AND m.is_netflix = 1)
        OR (:platform_name = 'Hotstar' AND m.is_hotstar = 1)
    """
    result = connection.execute(text(query), {"platform_name": platform_name})
    return result.fetchall()
```

```python
def get_top_rated_movies_by_genre(connection, genre_name, limit=10):
    query = """
    SELECT m.title, m.imdb_score, m.year AS release_year
    FROM movies m
    JOIN movie_genres mg ON m.id = mg.movie_id
    JOIN genres g ON mg.genre_id = g.id
    WHERE g.name = :genre_name
    ORDER BY m.imdb_score DESC
    LIMIT :limit
    """
    result = connection.execute(
        text(query),
        {"genre_name": genre_name, "limit": limit}
    )
    return result.fetchall()
```

```python
def get_movies_by_year_and_platform(connection, year, platform_name):
    query = """
    SELECT m.title, m.imdb_score, m.rating
    FROM movies m
    WHERE m.year = :year
      AND (
        (:platform_name = 'Amazon' AND m.is_amazon = 1)
        OR (:platform_name = 'Netflix' AND m.is_netflix = 1)
        OR (:platform_name = 'Hotstar' AND m.is_hotstar = 1)
      )
    """
    result = connection.execute(text(query), {"year": year, "platform_name": platform_name})
    return result.fetchall()
```

```python
def get_high_rated_movies_by_genre_and_platform(connection, genre_name, platform_name, min_imdb_score=8.0):
    query = """
    SELECT m.title, m.imdb_score, m.year AS release_year
    FROM movies m
    JOIN movie_genres mg ON m.id = mg.movie_id
    JOIN genres g ON mg.genre_id = g.id
    WHERE m.imdb_score > :min_imdb_score
      AND g.name = :genre_name
      AND (
        (:platform_name = 'Amazon' AND m.is_amazon = 1)
        OR (:platform_name = 'Netflix' AND m.is_netflix = 1)
        OR (:platform_name = 'Hotstar' AND m.is_hotstar = 1)
      )
    """
    result = connection.execute(text(query), {
        "min_imdb_score": min_imdb_score,
        "genre_name": genre_name,
        "platform_name": platform_name
    })
    return result.fetchall()
```

```python
def get_movies_by_cast_member(connection, cast_name):
    query = """
    SELECT m.title, m.year AS release_year
    FROM movies m
    JOIN movie_cast mc ON m.id = mc.movie_id
    JOIN cast c ON mc.cast_id = c.id
    WHERE c.name = :cast_name
    """
    result = connection.execute(text(query), {"cast_name": cast_name})
    return result.fetchall()
```

```python
def get_movies_by_imdb_score_range(connection, min_score, max_score):
    query = """
    SELECT title, imdb_score, year AS release_year
    FROM movies
    WHERE imdb_score BETWEEN :min_score AND :max_score
    ORDER BY imdb_score DESC
    """
    result = connection.execute(text(query), {"min_score": min_score, "max_score": max_score})
    return result.fetchall()
```

```python
def get_movies_by_year(connection, year):
    query = """
    SELECT title, imdb_score, runtime
    FROM movies
    WHERE year = :year
    """
    result = connection.execute(text(query), {"year": year})
    return result.fetchall()
```

```python
def get_movies_by_duration(connection, min_duration, max_duration):
    query = """
    SELECT title, runtime, year AS release_year
    FROM movies
    WHERE runtime BETWEEN :min_duration AND :max_duration
    ORDER BY runtime
    """
    result = connection.execute(text(query), {"min_duration": min_duration, "max_duration": max_duration})
    return result.fetchall()
```

```python
def get_movies_by_rating(connection, rating):
    query = """
    SELECT title, imdb_score, year AS release_year
    FROM movies
    WHERE rating = :rating
    """
    result = connection.execute(text(query), {"rating": rating})
    return result.fetchall()
```
